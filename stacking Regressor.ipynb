```python
import pandas as pd
import numpy as np
from sklearn.ensemble import StackingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
```

```python
df = pd.read_csv('insurance.csv')
```

```python
df.head()
```

```python
base_learner = [('clf1',DecisionTreeRegressor('mse')),('clf2',KNeighborsRegressor(n_neighbors=3))]
```

```python
algo = LinearRegression()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
df=pd.get_dummies(df,columns = ['sex'], prefix = 'R')
```

```python
df=pd.get_dummies(df,columns = ['smoker'], prefix = 'R')
```

```python
df['region'] = encoder.fit_transform(df['region'] )
```

```python
df
```

```python
x= df.drop('charges', axis =1)
x
```

```python
y = df[['charges']]
```

```python
y
```

```python
from sklearn.ensemble import StackingRegressor
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y, train_size = 0.8)
```

```python
stacking = StackingRegressor(estimators =base_learner, final_estimator= algo)
```
